```python
x = df.drop(columns=['price'], axis=1)
y = df['price']
```

```python
x
```

```python
# one hot encoding for categorical variables
x_one_encode = pd.get_dummies(x, columns=['transmission', 'fuelType', 'model'], drop_first=True)
```

```python
x_one_encode = x_one_encode.astype(int)
x_one_encode
```

```python
# level encoding for categorical variables
%pip install scikit-learn
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
columns = ['transmission', 'fuelType', 'model']
xlabel = x
for i in columns:
    xlabel[i] = encode.fit_transform(xlabel[i])
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['year', 'mileage', 'engineSize', 'tax', 'mpg']
scaler = StandardScaler()
x_one_encode[ numerical_cols] = scaler.fit_transform(x_one_encode[numerical_cols])
```

```python
x_one_encode
```

```python
x.columns
```

```python
x[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']] = scaler.fit_transform(
    x[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']]
)
```

```python
x
```

#### Training for model first one hot encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train_train, X_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
adj_r2
```
